## 4. Sum of Current & Previous Number

```
previous = 0
for current in range(1, 11):
    total = current + previous
    print("Current Number", current, "Previous Number", previous, "Sum:", total)
    previous = current
```

```
nums = [10, 20, 33, 46, 55]
for n in nums:
    if n % 5 == 0:
        print(n)
```

## 6. Prime Number Check

```
num = int(input("Enter the number: "))
is_prime = True

if num < 2:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print(num, "is a prime number.")
else:
    print(num, "is not a prime number.")
```

## 7. Reverse a List

```
lst = [10, 40, 30, 70]
rev = []
for i in range(len(lst)-1, -1, -1):
    rev.append(lst[i])
print(rev)
```

## 10. Diamond Pattern

```
for i in range(1, 6):
    print("*" * i)
for i in range(4, 0, -1):
    print("*" * i)
```

# Set 2 – Python Basics Part 2

## 1. Character Count

```
s = input("Enter a string value: ")
for ch in sorted(set(s)):
    print(ch, "=", s.count(ch))
```

## 2. Maximum of Three

```
def maximum(a, b, c):
    return max(a, b, c)

print(maximum(34, 12, 7))
```

## 3. Exponent Function

```
def exponent(base, exp):
    return base ** exp

print(exponent(2, 3))
```

## 4. Sum of Cubes

```
n = int(input("Enter number: "))
total = sum(i**3 for i in range(1, n))
print(total)
```

## 5. FizzBuzz 1–10

```
for i in range(1, 11):
    if i % 2 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 2 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

## 6. Most Frequent Item

```
nums = [2, 3, 4, 2, 5, 2]
print(max(nums, key=nums.count))
```

## 7. Sum of Squares

```
lst = [2, 1, 3, 1]
print(sum(i*i for i in lst))
```

## 8. Odd/Even 1–15

```
for i in range(1, 16):
    print(i, "- even" if i % 2 == 0 else "- odd")
```

## 9. Temperature Conversion

```
f = float(input("Temperature in Fahrenheit: "))
c = (f - 32) * 5/9
print("Temperature in Celsius =", c)
```

## 10. Factorial Function

```
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

print(factorial(3))
```

# Set 3 – Python Practice Problems

## 1. Sum of Digits

```
num = int(input("Enter number: "))
s = 0
for d in str(num):
    s += int(d)
print(s)
```

## 2. Factorial Function

```
def fact(n):
    f = 1
    for i in range(1, n+1):
        f *= i
    return f

print(fact(5))
```

## 3. Palindrome Check

```
s = input("Enter word: ").replace(" ", "").lower()
print("Palindrome" if s == s[::-1] else "Not Palindrome")
```

## 4. Vowel & Consonant Count

```
s = input("Enter sentence: ").lower()
vowels = "aeiou"
v = c = 0
for ch in s:
    if ch.isalpha():
        if ch in vowels:
            v += 1
        else:
            c += 1
print("Vowels:", v)
print("Consonants:", c)
```

## 5. Common Elements in Two Lists

```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
common = [x for x in a if x in b]
print(common)
```

## 6. Character Frequency

```
s = input("Enter string: ")
freq = {}
for ch in s:
    freq[ch] = freq.get(ch, 0) + 1
print(freq)
```

## 7. Leap Year Check

```
year = int(input("Enter year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not Leap Year")
```

## 8. Prime Check

```
n = int(input("Enter number: "))
if n < 2:
    print("Not Prime")
else:
    for i in range(2, n):
        if n % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
```

## 9. Fibonacci Sequence

```
t = int(input("Enter number of terms: "))
a, b = 0, 1
for _ in range(t):
    print(a, end=" ")
    a, b = b, a + b
```

## 10. Quadratic Equation Roots

```
import math

a = float(input("a: "))
b = float(input("b: "))
c = float(input("c: "))

d = b*b - 4*a*c

if d >= 0:
    r1 = (-b + math.sqrt(d)) / (2*a)
    r2 = (-b - math.sqrt(d)) / (2*a)
    print("Roots:", r1, "and", r2)
else:
    print("No real roots")
```
