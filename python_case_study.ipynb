```python
# Case Study 1: Bank Account Manager

balance = 0  # initial balance

def deposit(amount):
    global balance
    balance += amount
    print(f"Deposited â‚¹{amount}. New Balance: â‚¹{balance}")

def withdraw(amount):
    global balance
    if amount > balance:
        print("âŒ Withdrawal failed! Insufficient balance.")
    else:
        balance -= amount
        print(f"Withdrew â‚¹{amount}. New Balance: â‚¹{balance}")

def check_balance():
    print(f"Current Balance: â‚¹{balance}")

# ---- Test the functions ----
deposit(5000)
withdraw(2000)
withdraw(4000)
check_balance()
```

```python
# Case Study 2: Weather Data Analyzer

temperatures = [30, 32, 35, 28, 29, 31]  # Â°C
humidity_values = [70, 65, 80, 75, 60, 68]  # %

def average_temperature(data):
    return sum(data) / len(data)

def average_humidity(data):
    return sum(data) / len(data)

def weather_condition(temp, humidity):
    if temp > 33:
        print("Weather Condition: Hot ðŸŒž")
    elif temp < 25:
        print("Weather Condition: Cold â„ï¸")
    else:
        print("Weather Condition: Pleasant â›…")

    if humidity > 75:
        print("Humidity Status: Very Humid ðŸ’§")
    elif humidity < 40:
        print("Humidity Status: Dry ðŸŒµ")
    else:
        print("Humidity Status: Normal Air ðŸ’¨")

# ---- Calculate Results ----
avg_temp = average_temperature(temperatures)
avg_humid = average_humidity(humidity_values)

print(f"Average Temperature: {avg_temp:.2f} Â°C")
print(f"Average Humidity: {avg_humid:.2f}%")

# ---- Determine Weather ----
weather_condition(avg_temp, avg_humid)
```
