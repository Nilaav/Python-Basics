## Q1: Write a function that returns the range (maximum and minimum) of a given list of numbers.

```python
def find_range(numbers):
    if not numbers:
        return None
    return min(numbers), max(numbers)
```

```python
find_range([10, 2, 30, 7])
```

## Q2: Write a function that returns the mean value of a list of numbers.

```python
def mean(numbers):
    return sum(numbers) / len(numbers)
```

```python
mean([5, 10, 15])
```

## Q3: Write a function that returns the median value of a list of numbers.

```python
def median(numbers):
    n = len(numbers)
    numbers = sorted(numbers)
    mid = n // 2
    if n % 2 == 1:
        return numbers[mid]
    else:
        return (numbers[mid - 1] + numbers[mid]) / 2
```

```python
median([7, 2, 5, 4])
```

## Q4: Write a function that returns the mode of a list of numbers. If no mode exists, return None. If more than one mode exists, return the modes as a list.

```python
def mode(numbers):
    freq = {}
    for n in numbers:
        freq[n] = freq.get(n, 0) + 1
    max_count = max(freq.values())
    if max_count == 1:
        return None
    modes = [k for k, v in freq.items() if v == max_count]
    return modes if len(modes) > 1 else modes[0]
```

```python
mode([1, 2, 2, 3, 4])
```

## Q5: Write a function to compute a given percentile value (e.g., 90th percentile) of a list of numbers.

```python
def percentile(numbers, p):
    numbers = sorted(numbers)
    index = (p / 100) * (len(numbers) - 1)
    lower = int(index)
    upper = lower + 1
    if upper >= len(numbers):
        return numbers[lower]
    fraction = index - lower
    return numbers[lower] + fraction * (numbers[upper] - numbers[lower])
```

```python
percentile([10, 20, 30, 40, 50], 90)
```

## Q6: Write a function that returns the lower bound of a list of numbers.

```python
def lower_bound(numbers):
    numbers = sorted(numbers)
    def med(nums):
        l = len(nums)
        m = l // 2
        if l % 2 == 1:
            return nums[m]
        else:
            return (nums[m - 1] + nums[m]) / 2
    n = len(numbers)
    mid = n // 2
    q1 = med(numbers[:mid])
    q3 = med(numbers[mid:] if n % 2 == 0 else numbers[mid+1:])
    iqr = q3 - q1
    return q1 - 1.5 * iqr
```

```python
lower_bound([10, 12, 15, 18, 100])
```

## Q7: Write a function to return any outliers from a given list of numbers. If no outliers are present, return None.

```python
def outliers(numbers):
    numbers = sorted(numbers)
    def med(nums):
        l = len(nums)
        m = l // 2
        if l % 2 == 1:
            return nums[m]
        else:
            return (nums[m - 1] + nums[m]) / 2
    n = len(numbers)
    mid = n // 2
    q1 = med(numbers[:mid])
    q3 = med(numbers[mid:] if n % 2 == 0 else numbers[mid+1:])
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers_list = [x for x in numbers if x < lower or x > upper]
    return outliers_list if outliers_list else None
```

```python
outliers([10, 12, 15, 18, 100])
```

## Q8: Write a function that returns the standard deviation of a list of numbers.

```python
def standard_deviation(numbers):
    m = sum(numbers)/len(numbers)
    variance = sum((x - m) ** 2 for x in numbers) / len(numbers)
    return variance ** 0.5
```

```python
standard_deviation([10, 20, 30])
```

## Q9: Write a function to compute and return Z-score values of a list of numbers.

```python
def z_scores(numbers):
    m = sum(numbers)/len(numbers)
    sd = (sum((x - m) ** 2 for x in numbers) / len(numbers)) ** 0.5
    return [(x - m) / sd for x in numbers]
```

```python
z_scores([10, 20, 30])
```

## Q10: Write a function to compute and return the Pearson correlation coefficient value of two lists of numbers.

```python
def pearson_correlation(x, y):
    if len(x) != len(y):
        return None
    mean_x = sum(x)/len(x)
    mean_y = sum(y)/len(y)
    num = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(len(x)))
    den_x = sum((xi - mean_x) ** 2 for xi in x)
    den_y = sum((yi - mean_y) ** 2 for yi in y)
    denominator = (den_x * den_y) ** 0.5
    if denominator == 0:
        return None
    return num / denominator
```

```python
pearson_correlation([1,2,3,4], [10,20,30,40])
```
